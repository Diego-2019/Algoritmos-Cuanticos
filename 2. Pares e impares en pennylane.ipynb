# Pares e impares en pennylane

Crea un algoritmo donde se visualice la probabilidad de obtener numeros pares e impares en pennylane.

## Pares






```python
import pennylane as qml
from pennylane import numpy as np
import matplotlib.pyplot as plt
from functools import partial
```

```python
# Define the quantum device
dev = qml.device("default.qubit", wires=3)
```

```python
# Define the quantum circuit
@partial(qml.set_shots, shots=1000)
@qml.qnode(dev)
def circuit1():
    qml.Hadamard(wires=0)
    qml.Hadamard(wires=1)
    return qml.probs(wires=[0, 1, 2])
```

```python
# Draw the circuit using matplotlib
qml.drawer.use_style("black_white_dark")
fig, ax = qml.draw_mpl(circuit1)()
plt.show()
```

```python
# Print the output probabilities
result = circuit1()

sum = 0
for i, prob in enumerate(result):
    print(f"Probability of state |{i:03b}>: {prob:.4f}")
    if prob > 0:
        sum += prob

print(f"Suma de probabilidades: {sum}")

```

## Impares

```python
# Define the quantum circuit
@partial(qml.set_shots, shots=1000)
@qml.qnode(dev)
def circuit2():
    qml.Hadamard(wires=0)
    qml.Hadamard(wires=1)
    qml.PauliX(wires=2)
    return qml.probs(wires=[0, 1, 2])
```

```python
# draw the circuit using matplotlib
qml.drawer.use_style("black_white_dark")
fig, ax = qml.draw_mpl(circuit2)()
plt.show()
```

```python
result = circuit2()

sum = 0
for i, prob in enumerate(result):
    print(f"Probability of state |{i:03b}>: {prob:.4f}")
    if prob > 0:
        sum += prob

print(f"Suma de probabilidades: {sum}")
```

## Pares e impares (correcci√≥n)

```python
# Define the device
dev = qml.device('default.qubit', wires=3)

# Define the quantum circuit
@qml.qnode(dev)
def circuit():
    for i in range(3):
        qml.Hadamard(wires=i)
    return qml.probs(wires=2)

# Draw the circuit using matplotlib
qml.drawer.use_style("black_white_dark")
fig, ax = qml.draw_mpl(circuit)()
plt.show()
```

```python
result = circuit()
print(result)
print(f"Probability of measuring an even number: {result[0]:.1f}")
print(f"Probability of measuring an odd number: {result[1]:.1f}")
```
