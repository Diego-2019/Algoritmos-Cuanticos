<h1 align="center">Estados de Bell</h1>
Realiza los circuitos cuánticos para los 4 estados de Bell en Pennylane

-Dibuja el circuito 

-Muestra la salida como un vector de estado.

```python
import pennylane as qml
from pennylane import numpy as np
import matplotlib.pyplot as plt
# from functools import partial
```

```python
# Define the quantum device
dev = qml.device("default.qubit", wires=2)
```

## Estado $|\beta_{00} \rangle = \frac{1}{\sqrt{2}} (|00 \rangle + |11 \rangle) $

```python
# Define the quantum circuit
@qml.qnode(dev)
def bell00():
    # Create the Bell state |β00⟩
    qml.Hadamard(wires=0)
    qml.CNOT(wires=[0, 1])
    return qml.state()
```

```python
# draw the circuit
qml.drawer.use_style("black_white_dark")
fig, ax = qml.draw_mpl(bell00)()
plt.show()
```

```python
result = bell00()
print(f"Vector de estado: {result.round(3)}")
```

## Estado $|\beta_{01} \rangle = \frac{1}{\sqrt{2}} (|01 \rangle + |10 \rangle) $

```python
# define the quantum circuit
@qml.qnode(dev)
def bell01():
    # Create the Bell state |β01⟩
    qml.Hadamard(0)
    qml.PauliX(1)
    qml.CNOT([0, 1])
    return qml.state()
```

```python
# draw the circuit
qml.drawer.style.use_style("black_white_dark")
fig, ax = qml.draw_mpl(bell01)()
plt.show()
```

```python
result = bell01()
print(f"Vector de estado: {result.round(3)}")
```

## Estado $|\beta_{10} \rangle = \frac{1}{\sqrt{2}} (|00 \rangle - |11 \rangle) $

```python
# Define the quantum circuit
@qml.qnode(dev)
def bell10():
    # Create the Bell state |β10⟩
    qml.Hadamard(0)
    qml.CNOT([0, 1])
    qml.PauliZ(1)
    return qml.state()
```

```python
# Draw the circuit
qml.drawer.style.use_style("black_white_dark")
fig, ax = qml.draw_mpl(bell10)()
plt.show()
```

```python
result = bell10()
print(f"Vector de estado: {result.round(3)}")
```

## Estado $|\beta_{11} \rangle = \frac{1}{\sqrt{2}} (|01 \rangle - |10 \rangle) $

```python
# Define the circuit
@qml.qnode(dev)
def bell11():
    # Create the Bell state |β11⟩
    qml.Hadamard(0)
    qml.PauliX(1)
    qml.CNOT([0, 1])
    qml.PauliZ(0)
    return qml.state()
```

```python
# draw the circuit
qml.drawer.style.use_style("black_white_dark")
fig, ax = qml.draw_mpl(bell11)()
plt.show()
```

```python
result = bell11()
print(f"Vector de estado: {result.round(3)}")
```
