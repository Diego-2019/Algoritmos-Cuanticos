<h1 align="center">Codificación en Angulo</h1>
Input: N elementos random

Output: Codificación en angulo

```python
import pennylane as qml
import numpy as np
import matplotlib.pyplot as plt
```

```python
# ask for the number of elements
n = int(input("Enter the number of elements: "))
```

```python
np.random.seed(3)

# array with n random elements
array = np.round(np.random.uniform(0, 2 * np.pi, size=n), 3)

print(array)

# define the device
dev = qml.device("default.qubit", wires=n)
```

```python
# define the circuit
@qml.qnode(dev)
def circuit(array):
    # encode the array
    for i in range(n):
        qml.RX(array[i], wires=i)

    return qml.state()

# draw the circuit
qml.drawer.use_style("black_white_dark")
fig, ax = qml.draw_mpl(circuit)(array)
plt.show()
```

```python
# run the circuit
result = circuit(array)

print(result.round(3))
```

```python
# check if the state vector is normalized
sum = 0
for e in result:
    sum += np.abs(e)**2

print(f"Suma de los elementos del vector: {sum.round(3)}")
```
