<h1 align="center">Deutsch-Jozsa para 3 qubits</h1>
Crea el algoritmo de deustch jozsa para 3 qubits

```python
from qiskit import *
from qiskit_aer import Aer
from qiskit.visualization import plot_histogram
```

```python
# Create a quantum circuit
qc = QuantumCircuit(3, 2)

qc.x(2)
qc.h([0, 1, 2])
qc.barrier()
# Oracle for a balanced function
qc.cx(0, 2)
qc.cx(1, 2)
qc.barrier()
qc.h([0, 1])
qc.barrier()

# Measure the first qubit
qc.measure(0, 1)
qc.measure(1, 0)

# Draw the circuit
qc.draw('mpl')
```

```python
# Define the simulator
backend = Aer.get_backend("statevector_simulator")

# Compile the circuit
compiled_circuit = transpile(qc, backend=backend)
# Execute the circuit on the simulator
job = backend.run(compiled_circuit, shots=1000)
# Get the results of the execution
result = job.result()
# Get the counts of the results
counts = result.get_counts(qc)
# Draw the histogram of the results
plot_histogram(counts)
```
