<h1 align="center">QFT para 3 qubits</h1>

```python
import pennylane as qml
import numpy as np
import matplotlib.pyplot as plt
```

```python
# define the device
dev = qml.device("default.qubit", wires=3)
```

```python
# define the circuit
@qml.qnode(dev)
def circuit():
    # Initialize the state
    qml.PauliX(0)
    qml.PauliX(1)

    # QFT implementation
    qml.Hadamard(2)
    qml.ControlledPhaseShift(np.pi / 2**2, [0, 2])
    qml.ControlledPhaseShift(np.pi / 2, [1, 2])
    qml.Hadamard(1)
    qml.ControlledPhaseShift(np.pi / 2, [0, 1])
    qml.Hadamard(0)

    # Swap qubits to reverse the order
    qml.SWAP([0, 2])

    return qml.state()
```

```python
# draw the circuit
qml.drawer.use_style("black_white_dark")
fig, ax = qml.draw_mpl(circuit)()
plt.show()
```

```python
# print the statevector
print(f"Vector de estado: {circuit().round(3)}")
```
